```python
#crime_data = pd.read_csv(root_path +'/interventionscitoyendo.csv')   #Note that the .csv was edited whereby é was replaced by e and à by a
crime_data = pd.read_csv('data/4_Crime/interventionscitoyendo.csv')   #Note that the .csv was edited whereby é was replaced by e and à by a
print(crime_data.shape)
crime_data.head()
```

```python
print(crime_data.isnull().sum())
#PDQ isn<t analyzed and thus do not need to drop
```

### Date

```python
crime_data['DATE'] = pd.to_datetime(crime_data['DATE'])
crime_data['Year'] = pd.DatetimeIndex(crime_data['DATE']).year
crime_data['Month'] = pd.DatetimeIndex(crime_data['DATE']). month
crime_data['Day'] = pd.DatetimeIndex(crime_data['DATE']).day
crime_data['Day_of_week'] = crime_data['DATE'].dt.dayofweek
crime_data.head()
```

```python
#drop 2019 since year is not complete
crime_data = crime_data[crime_data.Year != 2019]
```

```python
#data binning
#Monday=0, Sunday=6.

conditions = [
    (crime_data['Day_of_week'] >=0 ) & (crime_data['Day_of_week'] < 5 ),
    (crime_data['Day_of_week'] >= 5 )]
choices = ['weekDay', 'WeekEnd']
crime_data['weekDayEnd'] = np.select(conditions, choices)

crime_data.weekDayEnd.value_counts()
```

### CATEGORIE

```python
#crime_data['CATEGORIE'] = crime_data.CATEGORIE.factorize()[0]
crime_data['CATEGORIE_fact'], label = crime_data.CATEGORIE.factorize()
print(label)

#data binning
#Death
crime_data.CATEGORIE[crime_data.CATEGORIE == 'Infractions entrainant la mort'] = 'Mort'
#Vols
crime_data.CATEGORIE[crime_data.CATEGORIE == 'Vol de vehicule a moteur'] = 'Vol'
crime_data.CATEGORIE[crime_data.CATEGORIE == 'Vol dans / sur vehicule a moteur'] = 'Vol'
crime_data.CATEGORIE[crime_data.CATEGORIE == 'Vols qualifies'] = 'Vol'
#Other
crime_data.CATEGORIE[crime_data.CATEGORIE == 'Mefait'] = 'Autre'
crime_data.CATEGORIE[crime_data.CATEGORIE == 'Introduction'] = 'Autre'

crime_data.head()
```

```python
#crime_data = pd.get_dummies(crime_data, prefix=['CATEG_'] , columns=['CATEGORIE'])
crime_data['CATG_Vol'] = np.where(crime_data['CATEGORIE']=='Vol',1,0)
crime_data['CATG_Mort'] = np.where(crime_data['CATEGORIE']=='Mort',1,0)
crime_data['CATG_Autre'] = np.where(crime_data['CATEGORIE']=='Autre',1,0)
```

```python
crime_data.head()
```

### Location (LATITUDE, LONGITUDE)
-ignoring X and Y columns since redundant

```python
print(crime_data.shape)
#latitude and longitude of 1.0 don\t make sense
crime_data = crime_data[crime_data['LATITUDE'] != 1.0]
print(crime_data.shape)
```

```python
#Count unique lat/long combos
print(crime_data.groupby(['LATITUDE','LONGITUDE'])['DATE'].nunique().count())
```

```python
#Create new columns of grid lat/long
crime_data['LAT_3sigFig'] = crime_data.LATITUDE.round(3)
crime_data['LONG_3sigFig'] = crime_data.LONGITUDE.round(3)
print(crime_data.groupby(['LAT_3sigFig','LONG_3sigFig'])['DATE'].nunique().count())
crime_data.head()
```

```python
#Count unique lat/long combos
crime_data_group1 = crime_data.groupby(['LAT_3sigFig','LONG_3sigFig','Month','Year','QUART'])['CATEGORIE_fact'].count().sort_values(ascending=False).reset_index()
print(crime_data_group1.shape)
crime_data_group1.head()
```

```python
#Count unique lat/long combos
crime_data_group2 = crime_data.groupby(['LAT_3sigFig','LONG_3sigFig','Month','Year','QUART'])['CATG_Vol','CATG_Mort','CATG_Autre'].sum().sort_values(ascending=False, by='CATG_Autre').reset_index()
crime_data_group2.head()
```

```python
crime_data_group = pd.merge(crime_data_group1, crime_data_group2, on=['LAT_3sigFig','LONG_3sigFig','Month','Year','QUART'], how='inner')
crime_data_group.head()
```

```python
crime_data_group.to_csv('Cleaned_crime_data.csv')
#After export, delete first column since it has IDs and thus useless
```
